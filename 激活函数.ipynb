sigmod

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt


def sigmoid(x):
    """
    使用 TensorFlow 的 sigmoid 函数
    """
    return tf.sigmoid(tf.constant(x, dtype=tf.float32))


# 生成 -10 到 10 之间的 100 个等间距的数据点
x = np.linspace(-10, 10, 100)
y = sigmoid(x).numpy()


# 绘制 sigmoid 函数图像
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='Sigmoid Function')


# 添加标题和轴标签
plt.title('Sigmoid Function using TensorFlow')
plt.xlabel('x')
plt.ylabel('sigmoid(x)')


# 显示图例
plt.legend()


# 显示网格线
plt.grid(True)


# 显示图像
plt.show()
```

sigmoid激活函数

```python
x = np.linspace(-10,10,100)
y = tf.nn.sigmoid(x,y)
plt.plot(x,y)
plt.grid()
```

tanh激活函数

```python
x = np.linspace(-10,10,100)
y = tf.nn.tanh(x)
plt.plot(x,y)
plt.grid()
```
